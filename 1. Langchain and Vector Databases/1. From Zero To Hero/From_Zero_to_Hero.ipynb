```python
from langchain_community.llms import LlamaCpp
from langchain_core.callbacks import CallbackManager, StreamingStdOutCallbackHandler
from langchain_core.prompts import PromptTemplate

callback_manager = CallbackManager([StreamingStdOutCallbackHandler()])

llm = LlamaCpp(
    model_path="llama-2-7b.Q2_K.gguf",
    temperature=0,
    max_tokens=2000,
    top_p=1,
    callback_manager=callback_manager,
    verbose=True,  # Verbose is required to pass to the callback manager
)
```

```python
text = "Suggest a personalized workout routine for someone looking to improve cardiovascular endurance and prefers outdoor activities."
print(llm.invoke(text))
```

```python
from langchain.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser


prompt = PromptTemplate(
    input_variables = ["product"],
    template = "What is a good name for a company that makes {product}? Specify it"
)

llm_chain = prompt | llm | StrOutputParser()

print(llm_chain.invoke({"product": "eco-friendly water bottles"}))
```

```python
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferMemory

conversation = ConversationChain(
    llm = llm,
    verbose = True,
    memory = ConversationBufferMemory()
)

conversation.predict(input = "Tell me about yourself")
conversation.predict(input = "What can you do?")
conversation.predict(input = "How can you help me with data analysis")

print(conversation)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain.vectorstores import DeepLake
from langchain.text_splitter import RecursiveCharacterTextSplitter

embeddings = HuggingFaceEmbeddings()
```

```python
texts = [
    'Napolean Bonaparte was born in 15 August 1769',
    'Louis XIV was born in 5 September 1638'
]
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs = text_splitter.create_documents(texts)
```

```python
activeloop_org_id = 'sushankggghimire'
activeloop_dataset_name = 'zero_to_hero'
dataset_path = f"hub://{activeloop_org_id}/{activeloop_dataset_name}"
db = DeepLake(dataset_path=dataset_path, embedding=embeddings)

# db.add_documents(docs)

retriever = db.as_retriever()
```

```python
retrieval_chain = llm | retriever
```

```python
from langchain.agents import AgentExecutor, create_tool_calling_agent
from langchain_core.tools import tool
from langchain.agents.react.agent import create_react_agent
from langchain.agents import AgentType
from langchain import hub
from langchain.tools.retriever import create_retriever_tool

prompt = hub.pull("hwchase17/react")


@tool
def retriever_func(query: str) -> str:
    """
    Searches and returns info about people.
    """
    print("Retriever function called with query:", query)
    documents = retriever.get_relevant_documents(query)
    formatted_documents = "\n".join([doc.page_content for doc in documents])
    return formatted_documents

# Register the retriever_func as a tool
tools = [retriever_func]

agent = create_react_agent(
    tools = tools,
    llm = llm,
    prompt = prompt
)
agent_executor = AgentExecutor(agent=agent, tools=tools, handle_parsing_errors=True)
```

```python
from langchain_google_community.search import GoogleSearchAPIWrapper
from langchain_core.tools import Tool
from langchain.agents import create_self_ask_with_search_agent
from langchain_mistralai import ChatMistralAI

search = GoogleSearchAPIWrapper()

prompt = hub.pull('hwchase17/react')
model = ChatMistralAI()
tool = Tool(
    name="Intermediate Answer",
    description="Search Google for recent results.",
    func=search.run,
)

tools = [tool]

agent = create_react_agent(model, tools, prompt)
agent_exec = AgentExecutor(agent=agent, tools=tools, handle_parsing_errors=True, max_iterations=6)
```

```python
agent_exec.invoke({"input": "Is Bangladesh politically stable now that Sheikh Hasina fled the country?"})
```

```python
prompt = hub.pull('hwchase17/react')

tools = [
    Tool(
        name="Intermediate Answer",
        description="Search Google for recent results.",
        func=search.run,
),
    Tool(
        name = 'Summarizer',
        description="Create a summary of the text",
        func = summarize_chain.invoke
    )
]

agent = create_react_agent(
    llm, tools, prompt
)

agent_exec = AgentExecutor(agent=agent, tools=tools, handle_parsing_errors=True)
```
